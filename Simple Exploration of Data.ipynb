```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import json
import csv
from pandas import DataFrame

%matplotlib inline 
```

```python
train_path = f'{path}train.json'
validaton_path = f'{path}validation.json'
test_path = f'{path}test.json'
```

```python
with open(train_path) as d1: #first check if it's a valid json file or not
    train_json = json.load(d1)
with open(validaton_path) as d2: #first check if it's a valid json file or not
    validation_json = json.load(d2)
with open(test_path) as d3: #first check if it's a valid json file or not
    test_json = json.load(d3)
```

```python
my_dic_data = train_json
keys= my_dic_data.keys()
dict_you_want1={'my_items1':my_dic_data['annotations']for key in keys}
dict_you_want2={'my_items2':my_dic_data['images']for key in keys}
df=pd.DataFrame(dict_you_want1)
fd = pd.DataFrame(dict_you_want2)
df2=df['my_items1'].apply(pd.Series)
#print ("df2",df2.head())
fd2=fd['my_items2'].apply(pd.Series)
#print ("fd2",fd2.head())
train_data = pd.merge(df2, fd2, on='image_id', how='outer')
```

```python
train_data['url']
```

```python
train_data['url'] = train_data['url'].apply(lambda x: str(x[0]))
```
